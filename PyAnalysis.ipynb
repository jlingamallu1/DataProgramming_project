```python
state = dataframe.groupby('State')
#print year
state_illnesses = state.sum()
#print state_illnesses
state_illnesses = state_illnesses.sort_values(by ='Illnesses', ascending=False)
#month and illness 
state_illnesses=state_illnesses.drop(state_illnesses.columns[[0,1,3,4]], axis=1)
state_illnesses[:10].plot(kind = 'barh',subplots = True ,legend = False,figsize=(7,5),title='States With Top Number Of Reported Cases')
```

```python
def find_index(state):
    a=(state_illnesses.index.values).tolist()
    inde = a.index(state)
    return np.sqrt(state_illnesses.iloc[inde][0]*10)   #calculation to bring down high values of reported cases for map

state_illnesses_list=(state_illnesses.index.values).tolist() #list of all the states present in data
illnesses_reported_dict = { state_name : find_index(state_name) for state_name in state_illnesses_list }
del illnesses_reported_dict['Guam']
del illnesses_reported_dict['Republic of Palau']
```

```python
from mpl_toolkits.basemap import Basemap as Basemap
from matplotlib.colors import rgb2hex
from matplotlib.patches import Polygon
from matplotlib.patches import Patch

m = Basemap(llcrnrlon=-119,llcrnrlat=22,urcrnrlon=-64,urcrnrlat=49,
            projection='lcc',lat_1=33,lat_2=45,lon_0=-95)

shp_info = m.readshapefile('st99_d00','states',drawbounds=True)
print(shp_info)

# choose a color for each state based on number of cases reported.
colors={}
statenames=[]
cmap = plt.cm.hot # use 'hot' colormap
vmin = 0; vmax = 770 # set range.
for shapedict in m.states_info:
    statename = shapedict['NAME']
    # skip DC and Puerto Rico.
    if statename not in ['District of Columbia','Puerto Rico']:
        pop = illnesses_reported_dict[statename]
        # calling colormap with value between 0 and 1 returns
        # rgba value.  Invert color range (hot colors are high
        # illness cases reported)
        if (1.-np.sqrt((pop-vmin)/(vmax-vmin)) < 0.25):
            colors[statename] = cmap(0.33000000)[:3]
        elif (0.25 <= (1.-np.sqrt((pop-vmin)/(vmax-vmin))) < 0.5 ):
            colors[statename] = cmap(0.5555500)[:3]
        elif (0.5 <= (1.-np.sqrt((pop-vmin)/(vmax-vmin))) < 0.75 ):
            colors[statename] = cmap(0.877777)[:3]
        else:
            colors[statename] = cmap(0.955555)[:3]
      
    statenames.append(statename)
# cycle through state names, color each one.
ax = plt.gca() # get current axes instance

for nshape,seg in enumerate(m.states):
   
    # skip DC and Puerto Rico. - No Coordinates present in basemap
    if statenames[nshape] not in ['District of Columbia','Puerto Rico']:
        color = rgb2hex(colors[statenames[nshape]]) 
        poly = Polygon(seg,facecolor=color,edgecolor=color)
        ax.add_patch(poly)

plt.title('Filling State Polygons by Illness cases reported')
orange_patch = Patch(color='orange', label='Medium Effected states')
yellow_patch = Patch(color='yellow', label='Low Effected states')
red_patch = Patch(color='red', label='High Effected states')
plt.legend(handles=[red_patch,orange_patch,yellow_patch],loc=0,prop={'size':9})
plt.show()
```

```python
#Checking with Different kinds of weather conditions
print('Seasonal Analysis')
month = dataframe.groupby('Month')
seasons_illnesses = month.sum()
winter = seasons_illnesses.iloc[11][1] + seasons_illnesses.iloc[1][1] + seasons_illnesses.iloc[0][1]
spring = seasons_illnesses.iloc[2][1] + seasons_illnesses.iloc[3][1] + seasons_illnesses.iloc[4][1]
summer = seasons_illnesses.iloc[5][1] + seasons_illnesses.iloc[6][1] + seasons_illnesses.iloc[7][1]
fall = seasons_illnesses.iloc[8][1] + seasons_illnesses.iloc[9][1] + seasons_illnesses.iloc[10][1]
#print winter,spring,summer,fall
seasons = [winter,spring,summer,fall]

if(seasons.index(max(seasons))==0):
    explode = (0.1, 0, 0, 0)
elif(seasons.index(max(seasons))==1):
    explode = (0, 0.1, 0, 0)
elif(seasons.index(max(seasons))==2):
    explode = (0, 0, 0.1, 0)
else:
    explode = (0, 0, 0, 0.1)
    
labels = 'winter','spring','summer','fall'
colors = ['yellowgreen','lightskyblue', 'orange','gold']
#explode = (0, 0.1, 0, 0)  # only "explode" the 2nd slice (i.e. 'Spring')
plt.pie(seasons, explode=explode, labels=labels, colors=colors,
         autopct='%1.1f%%',shadow=True, startangle=90)
```

```python
#Final Report Generation Top affected states with affecting Viruses
def top_virus_instate(df_src,state):
    df_src = df_src[df_src['State'].isin([state])]
    df_src=df_src.groupby('Genus Species').sum()
    df_src=(df_src.drop(df_src.columns[[0,1,3,4]], axis=1)).sort_values(by ='Illnesses', ascending=False)
    print('State Name : "' +state +'"')
    print("V1 : "+df_src.index[0])
    print("V2 : "+df_src.index[1])
    print("V3 : "+df_src.index[2]+'\n')
    
def top_state_invirus(df_src,virus):
    print('Genus Specimen : ' +virus)
    print('S1 : '+df_src.index[0])
    print('S2 : '+df_src.index[1])
    print('S3 : '+df_src.index[2]+'\n')
    
from collections import Counter
def finding_mode(lst):
    counter = Counter(lst)
    _,val = counter.most_common(1)[0]
    return [x for x,y in counter.items() if y == val]
```

```python
print('\t\t\t\tReport 1')
print('----------------------------------------------------------------------')
print('Most Illenss cases reported States with top illness causing agents\n')
print('------Winter------')
print('------------------')
top_virus_instate(df_winter_src,df_winter_states.index[0])
top_virus_instate(df_winter_src,df_winter_states.index[1])

print('------Spring------')
print('------------------')
top_virus_instate(df_spring_src,df_spring_states.index[0])
top_virus_instate(df_spring_src,df_spring_states.index[1])

print('------Summer------')
print('------------------')
top_virus_instate(df_summer_src,df_summer_states.index[0])
top_virus_instate(df_summer_src,df_summer_states.index[1])

print('------Fall------')
print('------------------')
top_virus_instate(df_fall_src,df_fall_states.index[0])
top_virus_instate(df_fall_src,df_fall_states.index[1])

print('----------------------------------------------------------------------')

state_list = [df_winter_states.index[0],df_winter_states.index[1],df_winter_states.index[2],
              df_spring_states.index[0],df_spring_states.index[1],df_spring_states.index[2],
              df_summer_states.index[0],df_summer_states.index[1],df_summer_states.index[2],
              df_fall_states.index[0],df_fall_states.index[1],df_fall_states.index[2]]

recurring_states = finding_mode(state_list)
print ("States affected mostly by genus species in all seasons " )
for i in range(0,len(recurring_states)):
    print(str(i+1)+". " +recurring_states[i] )
```
